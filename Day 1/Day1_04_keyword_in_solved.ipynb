# Klíčové slovo `in`

Klíčové slovo `in` v Pythonu slouží k:
- iteraci přes prvky kolekcí
- kontrole přítomnosti prvku v kolekci
- vyhledávání v řetězcích

## 1. Iterování přes seznamy

Pomocí `for` cyklu a klíčového slova `in` můžeme procházet všechny prvky seznamu.

```python
# Příklad: procházení seznamu zvířat
animals = ["cat", "dog", "mouse"]
for animal in animals:
    print(animal)
```

### Úloha 1: Doplň chybějící kód

Doplň cyklus, který vypíše všechna ovoce ze seznamu.

```python
fruits = ["jablko", "hruška", "banán", "pomeranč"]
# Řešení:
for fruit in fruits:
    print(fruit)
```

## 2. Iterování přes slovníky

Při iteraci přes slovník získáváme klíče. K hodnotám přistoupíme pomocí `slovnik[klic]`.

```python
# Příklad: procházení slovníku
animal = {
    "name": "cat",
    "breed": "moggy",
    "owner": "Baba Yaga"
}

for feature in animal:
    print(feature, animal[feature])
```

### Úloha 2: Oprav chybu v kódu

Následující kód obsahuje chybu. Najdi ji a oprav.

```python
kniha = {
    "název": "Hobit",
    "autor": "J.R.R. Tolkien",
    "rok": 1937
}

# Řešení: Chyba byla v tom, že se vypisoval celý slovník místo hodnoty
for vlastnost in kniha:
    print(vlastnost, kniha[vlastnost])
```

## 3. Kontrola přítomnosti prvku v seznamu

Klíčové slovo `in` vrací `True`, pokud je prvek v seznamu přítomen, jinak `False`.

```python
# Příklad: kontrola přítomnosti v seznamu
animals = ["cat", "dog", "mouse"]

print("dog" in animals)      # True
print("chicken" in animals)  # False
```

### Úloha 3: Vytvoř podmínku

Napiš kód, který zkontroluje, zda je "jahoda" v seznamu ovoce. Pokud ano, vypíše "Jahoda je v seznamu", jinak "Jahoda není v seznamu".

```python
ovoce = ["jablko", "hruška", "jahoda", "meloun"]
# Řešení:
if "jahoda" in ovoce:
    print("Jahoda je v seznamu")
else:
    print("Jahoda není v seznamu")
```

## 4. Kontrola přítomnosti klíče ve slovníku

Ve slovníku klíčové slovo `in` kontroluje, zda existuje daný klíč (ne hodnota!).

```python
# Příklad: kontrola klíčů ve slovníku
animal = {
    "name": "cat",
    "breed": "moggy",
    "owner": "Baba Yaga"
}

print("breed" in animal)  # True
print("color" in animal)  # False
```

### Úloha 4: Doplň kontrolu

Doplň kód tak, aby zkontroloval, zda slovník `student` obsahuje klíč "vek". Pokud ano, vypíše hodnotu, jinak vypíše "Věk není uveden".

```python
student = {
    "jméno": "Jan",
    "třída": "3.A"
}

# Řešení:
if "vek" in student:
    print(student["vek"])
else:
    print("Věk není uveden")
```

## 5. Vyhledávání v řetězcích

Klíčové slovo `in` lze použít k ověření, zda je jeden řetězec obsažen v druhém.

```python
# Příklad: vyhledávání v textu
phrase = "The quick brown fox jumps over the lazy dog"

print("fox" in phrase)  # True
print("dog" in phrase)  # True
```

### Úloha 5: Oprav chybu

Kód má chybu. Najdi ji a oprav.

```python
text = "Python je skvělý programovací jazyk"
# Řešení: Python musí být v uvozovkách jako řetězec
print("Python" in text)
```

### Úloha 7: Filtrování seznamu

Máš seznam čísel. Vytvoř nový seznam, který obsahuje pouze čísla větší než 10.

```python
cisla = [5, 12, 3, 18, 7, 21, 9, 15]
# Řešení:
velka_cisla = []
for cislo in cisla:
    if cislo > 10:
        velka_cisla.append(cislo)
print(velka_cisla)
```

### Úloha 8: Kontrola e-mailu

Napiš kód, který zkontroluje, zda řetězec obsahuje znak "@" a tečku ".". Pokud ano, vypíše "Platný e-mail", jinak "Neplatný e-mail".

```python
email = "student@skola.cz"
# Řešení:
if "@" in email and "." in email:
    print("Platný e-mail")
else:
    print("Neplatný e-mail")
```

### Úloha 9: Počítání výskytů

Napiš program, který spočítá, kolikrát se v seznamu `barvy` vyskytuje barva "červená".

```python
barvy = ["modrá", "červená", "zelená", "červená", "žlutá", "červená"]
# Řešení:
pocet = 0
for barva in barvy:
    if barva == "červená":
        pocet += 1
print("Barva 'červená' se vyskytuje", pocet, "krát")
```

### Úloha 10: Vytvoř menu

Vytvoř slovník `menu` s 5 jídly a jejich cenami. Napiš kód, který:
1. Vypíše všechna jídla s cenami
2. Zkontroluje, zda menu obsahuje "pizza"
3. Pokud ano, vypíše její cenu

```python
# Řešení:
menu = {
    "pizza": 150,
    "hamburger": 120,
    "salát": 80,
    "polévka": 50,
    "dezert": 70
}

# 1. Výpis všech jídel s cenami
print("Nabídka:")
for jidlo in menu:
    print(f"{jidlo}: {menu[jidlo]} Kč")

# 2. a 3. Kontrola a výpis ceny pizzy
if "pizza" in menu:
    print(f"\nPizza stojí {menu['pizza']} Kč")
else:
    print("\nPizza není v nabídce")
```
