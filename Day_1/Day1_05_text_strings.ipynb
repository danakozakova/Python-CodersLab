# Textové řetězce (strings)

## Co je to string?

String je textový řetězec - prostě jakýkoliv text.

String je cokoliv, co je v uvozovkách.

### Jednoduché a dvojité uvozovky

Pro vytvoření stringu můžete použít dvojité uvozovky:

```python
begin = ("The man in black fled across the desert," +
         " and the gunslinger followed.")
begin
```

Nebo jednoduché uvozovky (apostrofy). Pro Python je to jedno:

```python
title = 'Harry Potter and the Prisoner of Azkaban'
title
```

### Víceřádkové stringy

Pro víceřádkový text použijeme trojité uvozovky:

```python
screenplay = """
VADER
If only you knew the power of the Dark Side. Obi-Wan never told you what happened
to your father.

LUKE
He told me enough! He told me you killed him!

VADER
No, I am your father.

LUKE - (shocked)
No. No! That's not true! That's impossible!
"""

screenplay
```

```python
print(screenplay)
```

## Formátování stringů

### Spojování stringů operátorem +

Stringy lze spojovat operátorem `+`:

```python
print("Bilbo" + "Baggins")
```

Pozor - mezi slova se mezera nedoplní automaticky!

### Úloha 3: Oprav výstup

Uprav kód tak, aby mezi jménem a příjmením byla mezera:

```python
jmeno = "Bilbo"
prijmeni = "Baggins"
```

```python
# Uprav:
cele_jmeno = jmeno + prijmeni
print(cele_jmeno)
```

### Formátování pomocí %

Starší způsob formátování - podobný jako v jazyce C:

```python
hobbits = ("Frodo", "Sam", "Merry", "Pippin")

desc = ("Hobbits in LOTR: %s, %s, %s" +
        " and %s.") % hobbits

print(desc)
```

Důležité pravidlo:
Počet %s musí odpovídat počtu prvků v tuple!
* 4× %s = potřebujeme 4 prvky
* Pokud by bylo méně/více, dostanete chybu

### Metoda format()

Modernější způsob formátování:

```python
description = "{} is a {}.".format("Gandalf The Grey", "wizard")

print(description)
```

S pojmenovanými parametry:

```python
description = "{name} je {occupation}.".format(
    name="Fero",
    occupation="obchodník"
)

print(description)
```

### F-stringy (od Pythonu 3.6)

Nejmodernější a nejčitelnější způsob formátování:

```python
name = "Han Solo"
occupation = "smuggler"

description = f"{name} is a {occupation}."

print(description)
```
