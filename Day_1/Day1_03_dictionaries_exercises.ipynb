## Cvičení 1: Uživatelská data

Napiš funkci `uzivatelska_data(jmeno, prijmeni)`, která přijímá jako argumenty jméno a příjmení a vrací slovník s následujícími klíči a hodnotami:
- `"jmeno"` - hodnota z argumentu jméno,
- `"prijmeni"` - hodnota z argumentu příjmení,
- `"cele_jmeno"` - složené z obou argumentů oddělených mezerou,
- `"inicialy"` - první písmena obou argumentů následovaná tečkami; obě části oddělené mezerou.

**Příklad:**
```python
uzivatelska_data('Jan', 'Novák')
```

**Výsledek:**
```python
{"jmeno": "Jan", "prijmeni": "Novák", "cele_jmeno": "Jan Novák", "inicialy": "J. N."}
```

````python
# Zde napiš funkci "uzivatelska_data"
````

### Testování funkce

````python
if uzivatelska_data('Jan', 'Novák') == {
        "jmeno": "Jan",
        "prijmeni": "Novák",
        "cele_jmeno": "Jan Novák",
        "inicialy": "J. N."
}:
    print(' Test prošel!')
else:
    print('Funkce nevrátila očekávaný výsledek. Očekáváno:')
    print('{"jmeno": "Jan", "prijmeni": "Novák", "cele_jmeno": "Jan Novák", "inicialy": "J. N."}')
    print('Funkce vrátila:')
    print(uzivatelska_data('Jan', 'Novák'))

if uzivatelska_data('Anna', 'Svobodová') == {
        "jmeno": "Anna",
        "prijmeni": "Svobodová",
        "cele_jmeno": "Anna Svobodová",
        "inicialy": "A. S."
}:
    print(' Test prošel!')
else:
    print(' Funkce nevrátila očekávaný výsledek. Očekáváno:')
    print('{"jmeno": "Anna", "prijmeni": "Svobodová", "cele_jmeno": "Anna Svobodová", "inicialy": "A. S."}')
    print('Funkce vrátila:')
    print(uzivatelska_data('Anna', 'Svobodová'))
````

## Cvičení 2: Zobrazení slovníku

Napiš funkci `zobraz_slovnik(slovnik)`, která přijímá slovník jako argument a zobrazí ho způsobem uvedeným níže.

**Příklad:**
```python
zobraz_slovnik({'kocka': 'Kočka', 'pes': 'Pes', 'ptak': 'Pták'})
```

**Výstup:**
```
kocka=>Kočka
pes=>Pes
ptak=>Pták
```

**Úkol:** Použij cyklus `for` k procházení slovníku a vypíši každou dvojici klíč-hodnota na nový řádek.

````python
# Zde napiš funkci "zobraz_slovnik"
````

### Testování funkce

````python
import sys
import io
_real_stdout = sys.stdout

sys.stdout = io.StringIO()
zobraz_slovnik({'kocka': 'Kočka', 'pes': 'Pes'})
sys.stdout.seek(0)
if set(sys.stdout.read().strip().split()) == {'kocka=>Kočka', 'pes=>Pes'}:
    print(' Test prošel!', file=_real_stdout)
else:
    print(' Funkce nezobrazila očekávané řádky', file=_real_stdout)
    print('Funkce zobrazila:', file=_real_stdout)
    sys.stdout.seek(0)
    print(sys.stdout.read(), file=_real_stdout)
    print('Místo očekávaného:\nkocka=>Kočka\npes=>Pes\n', file=_real_stdout)

sys.stdout = io.StringIO()
zobraz_slovnik({'ptak': 'Pták', 'mys': 'Myš'})
sys.stdout.seek(0)
if set(sys.stdout.read().strip().split()) == {'ptak=>Pták', 'mys=>Myš'}:
    print(' Test prošel!', file=_real_stdout)
else:
    print(' Funkce nezobrazila očekávané řádky', file=_real_stdout)
    print('Funkce zobrazila:', file=_real_stdout)
    sys.stdout.seek(0)
    print(sys.stdout.read(), file=_real_stdout)
    print('Místo očekávaného:\nptak=>Pták\nmys=>Myš\n', file=_real_stdout)

sys.stdout = _real_stdout
````
