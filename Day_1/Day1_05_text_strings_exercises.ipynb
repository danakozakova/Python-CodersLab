## Cvičení 1: Vytvoření emailové adresy

Napište funkci `create_email(first_name, last_name)`, která vygeneruje emailovou adresu v doméně `myamazingcompany.com`. Uživatelské jméno by mělo být vytvořeno z argumentů: první písmeno jména, tečka, příjmení - vše malými písmeny.

Příklad:
```python
print(create_email('John', 'Connor'))
```

Výsledek:
```
j.connor@myamazingcompany.com
```

````python
# Sem napište funkci "create_email"
````

### Test cvičení 1:

````python
if create_email('John', 'Connor') == 'j.connor@myamazingcompany.com':
    print('Passed!')
else:
    print('Pro argumenty "John" a "Connor" se očekává j.connor@myamazingcompany.com')
    print('Vaše funkce vrátila:', create_email('John', 'Connor'))
    
####################################################################

if create_email('Anna', 'Newman') == 'a.newman@myamazingcompany.com':
    print('Passed!')
else:
    print('Pro argumenty "Anna" a "Newman" se očekává a.newman@myamazingcompany.com')
    print('Vaše funkce vrátila:', create_email('Anna', 'Newman'))
````

## Cvičení 2: Vytvoření formátované adresy

print(create_address('John', 'Connor', 'Hayvenhurst Drive', 12, 'Van Nyus', '14329'))

print("Řádek 1\nŘádek 2")

````python
# Sem napište funkci "create_address"
````

### Test cvičení 2:

## Cvičení 3: Anonymizace emailu

Napište funkci `anonymize(email)`, která přijme emailovou adresu jako argument a vrátí text skládající se z:
- prvních tří znaků emailové adresy
- tří hvězdiček: `'***'`
- posledních pěti znaků emailové adresy

Pokud je adresa kratší než 10 znaků, funkce by měla vrátit text skládající se z:
- tří hvězdiček: `'***'`
- posledních pěti znaků emailové adresy

Příklad:
```python
anonymize('j.connor@gmail.com')
```

```
j.c***l.com
```

```python
anonymize('jco@o2.pl')
```

```
***o2.pl
```

````python
# Sem napište funkci "anonymize"
````

### Test cvičení 3:

````python
if anonymize('john.connor@microsoft.com') == 'joh***t.com':
    print('Passed!')
else:
    print('Pro argument "john.connor@microsoft.com" se očekává joh***t.com')
    print('Vaše funkce vrátila:', anonymize('john.connor@microsoft.com'))

if anonymize('jc@xd.eu') == '***xd.eu':
    print('Passed!')
else:
    print('Pro argument "jc@xd.eu" se očekává ***xd.eu')
    print('Vaše funkce vrátila:', anonymize('jc@xd.eu'))

if anonymize('abcd@xd.eu') == 'abc***xd.eu':
    print('Passed!')
else:
    print('Pro argument "abcd@xd.eu" se očekává abc***xd.eu')
    print('Vaše funkce vrátila:', anonymize('abcd@xd.eu'))

if anonymize('abc@xd.eu') == '***xd.eu':
    print('Passed!')
else:
    print('Pro argument "abc@xd.eu" se očekává ***xd.eu')
    print('Vaše funkce vrátila:', anonymize('abc@xd.eu'))
````

## Cvičení 4: Obrácení stringu

Napište funkci `reverse_str(text)`, která přijme string jako argument a vrátí ho v obráceném pořadí.

Příklad:
```python
reverse_str('Test')
```

```
tseT
```

````python
# Sem napište funkci "reverse_str"
````

### Test cvičení 4:

````python
if reverse_str('ABCDefg') == 'gfeDCBA':
    print('Passed!')
else:
    print('Pro argument "ABCDefg" se očekává gfeDCBA')
    print('Vaše funkce vrátila:', reverse_str('ABCDefg'))

if reverse_str('12345') == '54321':
    print('Passed!')
else:
    print('Pro argument "12345" se očekává 54321')
    print('Vaše funkce vrátila:', reverse_str('12345'))
````
