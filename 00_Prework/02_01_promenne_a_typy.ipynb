## Pravidla pro pojmenov√°n√≠ promƒõnn√Ωch

Ne v≈°echny n√°zvy promƒõnn√Ωch jsou v Pythonu povolen√©. Pojƒème si vyzkou≈°et, kter√© n√°zvy funguj√≠ a kter√© ne.

### üîí Rezervovan√° slova (keywords)

Python m√° **rezervovan√° slova**, kter√° maj√≠ speci√°ln√≠ v√Ωznam v jazyce. **Nelze je pou≈æ√≠t** jako n√°zvy promƒõnn√Ωch!

Seznam rezervovan√Ωch slov:

```
False      class      finally    is         return
None       continue   for        lambda     try
True       def        from       nonlocal   while
and        del        global     not        with
as         elif       if         or         yield
assert     else       import     pass
break      except     in         raise
```

Vyzkou≈°ejte si, co se stane, kdy≈æ zkus√≠te pou≈æ√≠t rezervovan√© slovo jako n√°zev promƒõnn√©:

````python
# Pokus o pou≈æit√≠ rezervovan√©ho slova 'for'
for = 10
````

````python
# Pokus o pou≈æit√≠ rezervovan√©ho slova 'if'
if = 5
````

````python
# Pokus o pou≈æit√≠ rezervovan√©ho slova 'class'
class = "Python"
````

**Chybov√° hl√°≈°ka:** Python v√°m ≈ôekne `SyntaxError: invalid syntax`, proto≈æe tato slova maj√≠ v Pythonu speci√°ln√≠ v√Ωznam a nelze je pou≈æ√≠t jako n√°zvy promƒõnn√Ωch.

### ‚ùå Neplatn√© n√°zvy promƒõnn√Ωch

N√°sleduj√≠c√≠ n√°zvy **NEFUNGUJ√ç** - vyzkou≈°ejte je a pod√≠vejte se na chybovou hl√°≈°ku:

````python
# N√°zev zaƒç√≠naj√≠c√≠ ƒç√≠slic√≠
1 = 5
````

````python
# N√°zev zaƒç√≠naj√≠c√≠ ƒç√≠slic√≠
1a = 10
````

````python
# N√°zev zaƒç√≠naj√≠c√≠ ƒç√≠slic√≠ a podtr≈æ√≠tkem
1_ = 20
````

**Pravidlo:** N√°zev promƒõnn√© **nesm√≠ zaƒç√≠nat ƒç√≠slic√≠**!

### ‚úÖ Platn√© n√°zvy promƒõnn√Ωch

Tyto n√°zvy **FUNGUJ√ç** spr√°vnƒõ:

````python
# Prost√Ω n√°zev
a = 100
print(a)
````

````python
# P√≠smeno s ƒç√≠slic√≠ (ƒç√≠slice nen√≠ na zaƒç√°tku)
a1 = 200
print(a1)
````

````python
# Podtr≈æ√≠tka jsou povolen√°
a_b_c___95 = 300
print(a_b_c___95)
````

````python
# N√°zev m≈Ø≈æe zaƒç√≠nat podtr≈æ√≠tkem
_abc = 400
print(_abc)
````

````python
# Podtr≈æ√≠tko, pak ƒç√≠slice
_1a = 500
print(_1a)
````

### üö´ ≈†patn√© n√°zvy promƒõnn√Ωch (i kdy≈æ jsou technicky povolen√©)

Nƒõkter√© n√°zvy jsou sice platn√©, ale **nejsou vhodn√©** pro psan√≠ ƒçist√©ho k√≥du:

````python
# ‚ùå Mezery nejsou povolen√©!
# number Of Users = 10  # Toto by zp≈Øsobilo chybu
````

````python
# ‚ùå Speci√°ln√≠ znaky jako %% nejsou povolen√©!
# tep%%por = 5  # Toto by zp≈Øsobilo chybu
````

````python
# ‚ö†Ô∏è Tento n√°zev je platn√Ω, ale ned√°v√° smysl
asdf = 42
print(asdf)
````

````python
# ‚ö†Ô∏è Diakritika nen√≠ dobr√Ω n√°pad (i kdy≈æ funguje)
u≈æivatel = "Jan"
print(u≈æivatel)
````

**Doporuƒçen√≠ pro dobr√© n√°zvy promƒõnn√Ωch:**
- Pou≈æ√≠vejte **smyslupln√© n√°zvy** (nap≈ô. `vek`, `jmeno`, `pocet_studentu`)
- Pi≈°te n√°zvy **anglicky** (v praxi se to ƒçasto vy≈æaduje)
- **Bez diakritiky**
- Pou≈æ√≠vejte **mal√° p√≠smena** a podtr≈æ√≠tka (nap≈ô. `user_count`)
- Vyhnƒõte se zkratk√°m, kter√© ned√°vaj√≠ smysl

## Konstanty

Konstanty jsou hodnoty, kter√© by se **nemƒõly mƒõnit**. V Pythonu je konvence pojmenovat konstanty **VELK√ùMI P√çSMENY**:

````python
# Konstanta - mƒõla by z≈Østat nemƒõnn√° po celou dobu programu
FINAL_ANSWER = 42
print(FINAL_ANSWER)
````

````python
# Dal≈°√≠ p≈ô√≠klady konstant
PI = 3.14159
MAX_STUDENTS = 30
SCHOOL_NAME = "Na≈°e ≈°kola"

print(PI)
print(MAX_STUDENTS)
print(SCHOOL_NAME)
````

**Pozn√°mka:** Python technicky **nedok√°≈æe zabr√°nit** zmƒõnƒõ konstanty, ale konvence VELK√ùCH P√çSMEN ≈ô√≠k√° program√°tor≈Øm: "Tuto hodnotu nemƒõ≈à!"

````python
# Python v√°m dovol√≠ zmƒõnit konstantu, ale NEMƒöLI byste to dƒõlat!
FINAL_ANSWER = 100  # ‚ö†Ô∏è Technicky mo≈æn√©, ale ≈†PATN√ù n√°pad
print(FINAL_ANSWER)
````
