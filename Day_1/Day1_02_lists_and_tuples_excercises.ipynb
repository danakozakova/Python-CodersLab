## Úloha 1: Vytvoření speciálního seznamu

Napiš funkci `create_list(a, b, c)`, která přijímá tři argumenty libovolného typu a vrací seznam, jehož prvky jsou hodnoty argumentů v pořadí: `a`, `b`, `c`, `b`, `a`.

**Příklad:**
```python
my_list = create_list(1, 2, 3)  # výsledek: [1, 2, 3, 2, 1]
```

````python
# Napiš funkci "create_list" zde
def create_list(a, b, c):
    return [a, b, c]
````

````python
create_list(1, 2, 3)
````

````python
def create_list(a, b, c):
    return [a, b, c, b, a]
````

````python
create_list(1, 2, 3)
````

````python
if create_list(7, 8, 9) == list(range(7, 10)) + [8] + [7]:
    print('Passed!')
else:
    print('For arguments 7, 8, 9 expected the function to return the list: [7, 8, 9, 8, 7]')
    print('Your function returned:', create_list(7, 8, 9))


if create_list('Huey', 'Dewey', 'Louie') == 'Huey.Dewey.Louie.Dewey.Huey'.split('.'):
    print('Passed!')
else:
    print(
        "'For arguments 'Huey', 'Dewey', 'Louie' expected the function to return the list: " 
        "['Huey', 'Dewey', 'Louie', 'Dewey', 'Huey']"
    )
    print("Your function returned:", create_list('Huey', 'Dewey', 'Louie'))
````

## Úloha 2: Histogram

Napiš funkci `histogram(list_of_integers)`, která projde každé číslo v seznamu a zobrazí počet znaků `#` rovný hodnotě čísla — na novém řádku pro každé číslo.

**Nápověda:** Chcesh-li vygenerovat opakující se text, můžeš ho vynásobit číslem typu `int`:
```python
"na" * 16 + " Batman!"
```

**Příklad:**
```python
histogram([3, 9, 1, 5])
```

**Výstup:**
```
###
#########
#
#####
```

````python
# Napiš funkci "histogram" zde
# print s násobením zopakuje vypsání znaku
print('#'*3)
````

````python
# zkusim v cyklu
seznam = [3, 9, 1, 5]

for c in seznam:
    print(c * '#')
````

````python
# cyklus dam do funkce
def histogram(seznam):
    for c in seznam:
        print(c * '#')

# poznamka: neni uplne štastné vypisovat ve funkciu printem..
````

````python
# zavolam funkci
histogram([1, 3, 9, 1, 5])
````

````python
# spojim predesle dve funkce
histogram(create_list(1, 3, 5))
````
