# Úvod do NumPy

NumPy (Numeric Python) je knihovna pro práci s velkými, vícerozměrnými datovými sadami. Je základem knihovny Pandas.

Hlavní výhody NumPy:
- Data mají jeden předem určený typ
- Zabírá méně místa v paměti
- Je výrazně rychlejší než běžné seznamy

## Import knihovny

```python
import numpy as np
```

## Vytvoření pole (array)

Základní datový typ v NumPy je `array`. Můžeme ho vytvořit z běžného seznamu.

```python
lst = [1, 2, 3]
```

```python
arr = np.array(lst)
```

```python
print(arr)
```

## Indexování

Přístup k prvkům je stejný jako u seznamu.

```python
print(lst[1])
```

```python
print(arr[1])
```

## Slicing (výřezy)

Všimněte si malého rozdílu ve výstupu - NumPy pole nemá čárky mezi prvky.

```python
print(lst[1:])
```

```python
print(arr[1:])
```

## Operace s poli

V NumPy můžeme snadno provádět matematické operace na všech prvcích najednou.

```python
arr * 5
```

U seznamu by stejná operace seznam 5x zopakovala:

```python
lst * 5
```

## Vícerozměrná pole (ndarray)

`ndarray` je vícerozměrná verze pole - v podstatě matice.

```python
x = np.array([
    [1, 2, 3],
    [4, 5, 6]])
```

```python
x
```

Ověření typu proměnné:

```python
type(x)
```

## Přístup k prvkům 2D pole

Prvek z prvního řádku a prvního sloupce (indexy začínají od 0):

```python
x[0, 0]
```

Celý první sloupec (všechny řádky, sloupec 0):

```python
x[:, 0]
```

## Shrnutí použitých funkcí a metod

| Funkce/syntaxe | Popis |
|----------------|-------|
| `import numpy as np` | Import knihovny NumPy s aliasem np |
| `np.array(seznam)` | Vytvoření NumPy pole ze seznamu |
| `pole[index]` | Přístup k prvku 1D pole |
| `pole[radek, sloupec]` | Přístup k prvku 2D pole |
| `pole[:, sloupec]` | Výběr celého sloupce |
| `pole[radek, :]` | Výběr celého řádku |
| `pole * cislo` | Násobení všech prvků konstantou |
| `pole1 + pole2` | Sčítání dvou polí po prvcích |
| `type(promenna)` | Zjištění typu proměnné |
