## Cvičení 1

Napiš funkci `multiply(a, b)`, která vrátí výsledek násobení dvou čísel, která jsou jí předána jako parametry.

```python
# Napiš funkci "nasobeni" zde
def multiply(a, b):
    return a * b
```

```python
# Testování funkce nasobeni
print('2 * 3 =', multiply(2, 3), '(expected result is 6)')
print('2.5 * 4 =', multiply(2.5, 4), '(expected result is 10.0)')
print('4 * 0.25 =', multiply(4, 0.25), '(expected result is 1.0)')
print('0.4 * 2.5 =', multiply(0.4, 2.5), '(expected result is 1.0)')
print('4 * 3 + 5 =', multiply(4, 3) + 5, '(expected result is 17)')
print('10 + 2 * 6 =', 10 + multiply(2, 6), '(expected result is 22)')
```

## Cvičení 2

Napiš funkci `power(zaklad, exponent)`, která přijímá dva parametry a vrací výsledek: parametr `zaklad` umocněný na `exponent`.

Napiš funkci `root(cislo, stupen)`, která vypočítá `stupen`-tou odmocninu z čísla `cislo`. Nechť je stupeň odmocniny standardně číslo `2`.

```python
# Napiš funkce "mocnina" a "odmocnina" zde
# mocnina ma operator **
# n-ta odmocnina je vlastne mocnina 1/n

def power(zaklad, exponent):
    return zaklad ** exponent
```

```python
# Testování funkcí mocnina a odmocnina
print('2 ** 3 =', power(2, 3), '(expected result is 8)')
print('1.5 ** 2 =', power(1.5, 2), '(expected result is 2.25)')
print('2 ** 5 - 30 =', power(2, 5) - 30, '(expected result is 2)')
```

```python
def root(cislo, stupen = 2):
    return cislo ** (1/stupen)
```

```python
print('2nd root of 64 is', root(64), '(expected result is 8.0)')
print('2nd root of 64 is', root(64, 2), '(expected result is 8.0)')
print('4th root of 2401 with 5 subtracted is', root(2401, 4) - 5, '(expected result is 2)')
```

## Cvičení 3

Napiš funkci `usd_na_czk(usd)`, která přijímá jeden parametr: částku v dolarech (typ: `int` nebo `float`) a vrací částku v CZK.

Použij směnný kurz `1 USD = 23 CZK`.

```python
# Napiš funkci "usd_na_czk" zde
def usd_na_czk(usd):
    kurz = 23
    return usd * kurz
```

```python
# Testování funkce usd_na_czk
print('100 USD =', usd_na_czk(100), 'CZK (očekávaný výsledek je 2300 CZK)')
print('1.55 USD =', usd_na_czk(1.55), 'CZK (očekávaný výsledek je 35.65 CZK)')
print('4 USD - 20 CZK =', usd_na_czk(4) - 20, 'CZK (očekávaný výsledek je 72 CZK)')
```

## Cvičení 4

Napiš funkci `je_sude(cislo)`, která přijímá číslo a vrací `True`, pokud je číslo sudé, a `False`, pokud není.

Napiš funkci `je_liche(cislo)`, která přijímá číslo a vrací `True`, pokud je číslo liché, a `False`, pokud není. Můžeš vypočítat výsledek (jako ve funkci `je_sude`) nebo použít funkci `je_sude` v kódu funkce `je_liche`.

```python
24 % 2
```

```python
# Napiš funkce "je_sude" a "je_liche" zde
def je_sude(cislo):
    if cislo % 2 == 0: return True
    else: return False

def je_liche(cislo):
    return (not je_sude(cislo))
```

```python
# Testování funkcí je_sude a je_liche
if je_sude(4):
    print('Číslo 4 je sudé')
else:
    print('Číslo 4 je... liché? Nebo ne?')

if je_sude(7):
    print('Číslo 7 je... sudé? Nebo ne?')
else:
    print('Číslo 7 je liché')

if je_liche(9):
    print('Číslo 9 je liché')
else:
    print('Číslo 9 je... sudé? Nebo ne?')

if je_liche(12):
    print('Číslo 12 je... liché? Nebo ne?')
else:
    print('Číslo 12 je sudé')
```
