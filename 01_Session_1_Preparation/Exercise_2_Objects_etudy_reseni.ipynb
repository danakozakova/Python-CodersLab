# Cvičení 2: Práce se slovníky - Spotřeba paliva

## Zadání úlohy

Níže je kód, který obsahuje informace o spotřebě paliva různých motorů. Vaším úkolem je vypočítat, kolik paliva **v průměru** spotřebují benzínové motory a kolik naftové motory. Výsledky zobrazte pomocí funkce `print()`.

### Data

Máme k dispozici slovník `fuel_consumption` se spotřebou v litrech na 100 km.

```python
fuel_consumption = {  # v litrech/100km
  'Citroen': {
    'Berlingo': {
      'petrol_engine': 8.25,
      'diesel_engine': 5.75,
    },
    'C3': {
      'petrol_engine': 7.0,
      'diesel_engine': 5.0,
    },
  },
}
```

```python
fuel_consumption['Citroen']
```

```python
fuel_consumption['Citroen']['C3']
```

```python
fuel_consumption['Citroen']['C3']['diesel_engine']
```

## Ukázka: Jak přistupovat k datům

Nejdříve si ukážeme, jak můžeme přistupovat k jednotlivým hodnotám ve slovníku:

```python
# Příklad: Vypíšeme spotřebu benzínového motoru Citroen Berlingo
print('Benzínový motor Citroen Berlingo spotřebuje', fuel_consumption['Citroen']['Berlingo']['petrol_engine'], 'litrů/100km')
```

```python
# Příklad: Vypíšeme spotřebu benzínového motoru Citroen Berlingo
print(f"Benzínový motor Citroen Berlingo spotřebuje {fuel_consumption['Citroen']['Berlingo']['petrol_engine']} litrů/100km")
```

---

## Krok 1: Získání všech hodnot pro benzínové motory

**Úkol:** Vytvořte proměnnou `benzin_berlingo` a uložte do ní spotřebu benzínového motoru Citroen Berlingo.

```python
# Vaše řešení:
benzin_berlingo = fuel_consumption['Citroen']['Berlingo']['petrol_engine']

print('Benzin Berlingo:', benzin_berlingo)
```

**Úkol:** Vytvořte proměnnou `benzin_c3` a uložte do ní spotřebu benzínového motoru Citroen C3.

```python
# Vaše řešení:
benzin_c3 = fuel_consumption['Citroen']['C3']['petrol_engine']

print('Benzin C3:', benzin_c3)
```

---

## Krok 2: Výpočet průměrné spotřeby benzínových motorů

**Úkol:** Vypočítejte průměrnou spotřebu benzínových motorů. 

**Nápověda:** Průměr = (hodnota1 + hodnota2) / počet hodnot

```python
# Vaše řešení:
prumer_benzin = (benzin_berlingo + benzin_c3)/2

print('Průměrná spotřeba benzínových motorů:', prumer_benzin, 'l/100km')
```

---

## Krok 3: Získání všech hodnot pro naftové motory

**Úkol:** Vytvořte proměnné `nafta_berlingo` a `nafta_c3` a uložte do nich spotřebu naftových motorů.

```python
# Vaše řešení:
nafta_berlingo = fuel_consumption['Citroen']['Berlingo']['diesel_engine']
nafta_c3 = fuel_consumption['Citroen']['C3']['diesel_engine']

print('Nafta Berlingo:', nafta_berlingo)
print('Nafta C3:', nafta_c3)
```

---

## Krok 4: Výpočet průměrné spotřeby naftových motorů

**Úkol:** Vypočítejte průměrnou spotřebu naftových motorů.

```python
# Vaše řešení:
prumer_nafta = (nafta_berlingo + nafta_c3)/2

print('Průměrná spotřeba naftových motorů:', prumer_nafta, 'l/100km')
```

---

## Krok 5: Kompletní řešení v jedné buňce

**Úkol:** Teď zkuste napsat celé řešení najednou v jedné buňce. Vypočítejte oba průměry a vypište je.

```python
# Vaše kompletní řešení:

# 1. Získejte všechny hodnoty pro benzínové motory
benzin_berlingo = fuel_consumption['Citroen']['Berlingo']['petrol_engine']
benzin_c3 = fuel_consumption['Citroen']['C3']['petrol_engine']

# 2. Vypočítejte průměr pro benzínové motory
prumer_benzin = (benzin_berlingo + benzin_c3) / 2

# 3. Získejte všechny hodnoty pro naftové motory
nafta_berlingo = fuel_consumption['Citroen']['Berlingo']['diesel_engine']
nafta_c3 = fuel_consumption['Citroen']['C3']['diesel_engine']

# 4. Vypočítejte průměr pro naftové motory
prumer_nafta = (nafta_berlingo + nafta_c3) / 2

# 5. Vypište výsledky
print(f"Průměrná spotřeba benzínových motorů: {prumer_benzin} l/100km")
print(f"Průměrná spotřeba naftových motorů: {prumer_nafta} l/100km")
```

---

## Bonusová úloha 1: Který typ motoru je úspornější?

**Úkol:** Napište kód, který porovná průměrnou spotřebu benzínových a naftových motorů a vypíše, který typ je úspornější.

```python
# Vaše řešení:
# Nápověda: Použijte podmínku if/else
if prumer_benzin > prumer_nafta:
    print(f"Naftove motory jsou uspornejší. Rozdíl je: {prumer_benzin - prumer_nafta}")
elif prumer_nafta > prumer_benzin:
    print(f"Benzinové motory jsou uspornejší. Rozdíl je: {prumer_nafta - prumer_benzin}")
else: 
    print(f"Oba jsou rovnako usporné.")
```

## Bonusová úloha 2: Přidání nových dat

**Úkol:** Do slovníku `fuel_consumption` přidejte nový model auta (např. Citroen C4) s hodnotami pro benzínový a naftový motor. Poté přepočítejte průměry pro všechny tři modely.

**Nápověda:** Můžete použít hodnoty: benzínový motor 7.5 l/100km, naftový motor 5.5 l/100km

```python
# Vaše řešení:

# 1. Přidejte nový model do slovníku
fuel_consumption['Citroen']['C4'] = {
    'petrol_engine': 7.5,
    'diesel_engine': 5.5
}

# 2. Vypočítejte nové průměry (pro 3 modely)
benzin_c4 = fuel_consumption['Citroen']['C4']['petrol_engine']
nafta_c4 = fuel_consumption['Citroen']['C4']['diesel_engine']

prumer_benzin_3 = (benzin_berlingo + benzin_c3 + benzin_c4) / 3
prumer_nafta_3 = (nafta_berlingo + nafta_c3 + nafta_c4) / 3

# 3. Vypište výsledky
print(f'Průměrná spotřeba benzínových motorů (3 modely): {prumer_benzin_3:.2f} l/100km')
print(f'Průměrná spotřeba naftových motorů (3 modely): {prumer_nafta_3:.2f} l/100km')
```

---

## Bonusová úloha 3: Výpočet pomocí cyklu

**Úkol:** Pokud už umíte pracovat s cykly, zkuste vypočítat průměry pomocí cyklu `for`, který projde všechny modely.

**Nápověda:** Budete potřebovat sčítat hodnoty a počítat, kolik jich je.

```python
# Vaše řešení (pro pokročilé):

# Inicializace proměnných pro součty a počty
soucet_benzin = 0
soucet_nafta = 0
pocet_modelu = 0

# Projdeme všechny modely v slovníku
for model in fuel_consumption['Citroen']:
    soucet_benzin += fuel_consumption['Citroen'][model]['petrol_engine']
    soucet_nafta += fuel_consumption['Citroen'][model]['diesel_engine']
    pocet_modelu += 1

# Vypočítáme průměry
prumer_benzin_cyklus = soucet_benzin / pocet_modelu
prumer_nafta_cyklus = soucet_nafta / pocet_modelu

print(f'Průměrná spotřeba benzínových motorů (cyklus): {prumer_benzin_cyklus:.2f} l/100km')
print(f'Průměrná spotřeba naftových motorů (cyklus): {prumer_nafta_cyklus:.2f} l/100km')
```

## Alternativní řešení - vnořené cykly

```python
# Alternativní řešení s vnořenými cykly:
soucet_benzin = 0
soucet_nafta = 0
pocet_modelu = 0

for znacka in fuel_consumption:
    for model in fuel_consumption[znacka]:
        soucet_benzin += fuel_consumption[znacka][model]['petrol_engine']
        soucet_nafta += fuel_consumption[znacka][model]['diesel_engine']
        pocet_modelu += 1

prumer_benzin_vnorene = soucet_benzin / pocet_modelu
prumer_nafta_vnorene = soucet_nafta / pocet_modelu

print(f'Průměrná spotřeba benzínových motorů (vnořené cykly): {prumer_benzin_vnorene:.2f} l/100km')
print(f'Průměrná spotřeba naftových motorů (vnořené cykly): {prumer_nafta_vnorene:.2f} l/100km')
```
