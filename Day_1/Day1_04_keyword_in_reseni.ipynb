# Klíčové slovo `in`

Klíčové slovo `in` v Pythonu slouží k:
- iteraci přes prvky kolekcí
- kontrole přítomnosti prvku v kolekci
- vyhledávání v řetězcích

# ITEROVÁNÍ

## 1. Iterování přes seznamy

Pomocí `for` cyklu a klíčového slova `in` můžeme procházet všechny prvky seznamu.

```python
# Příklad: vypiš zvířata ze seznamu
animals = ["cat", "dog", "mouse"]

for zvire in animals:
    print(zvire)
```

## 2. Iterování přes slovníky

Při iteraci přes slovník získáváme klíče. K hodnotám přistoupíme pomocí `slovnik[klic]`.

```python
# Příklad: vypiš klíč a hodnotu ze slovníku
kniha = {
    "název": "Hobit",
    "autor": "J.R.R. Tolkien",
    "rok": 1937
}

for prvek in kniha:
    print(prvek)
```

```python
for index, prvek in kniha.items():
    print(index, prvek)
```

```python
for hodnota in kniha.values():
    print (hodnota)
```

# KONTROLA

## 3. Kontrola přítomnosti prvku v seznamu

Klíčové slovo `in` vrací `True`, pokud je prvek v seznamu přítomen, jinak `False`.

```python
# Příklad: kontrola přítomnosti v seznamu
animals = ["cat", "dog", "mouse"]
```

```python
"dog" in animals
```

```python
"chicken" in animals
```

Napiš kód, který zkontroluje, zda je "jahoda" v seznamu ovoce. Pokud ano, vypíše "Jahoda je v seznamu", jinak "Jahoda není v seznamu".

```python
ovoce = ["jablko", "hruška", "jahoda", "meloun"]
# Napiš svůj kód zde
```

```python
"jahoda" in ovoce
```

## 4. Kontrola přítomnosti klíče ve slovníku

Ve slovníku klíčové slovo `in` kontroluje, zda existuje daný **klíč (ne hodnota!)**.

```python
# Příklad: kontrola klíčů ve slovníku
animal = {
    "name": "cat",
    "breed": "moggy",
    "owner": "Baba Yaga"
}
```

```python
print("breed" in animal)
```

```python
print("cat" in animal)
```

Doplň kód tak, aby zkontroloval, zda slovník `student` obsahuje klíč "vek". Pokud ano, vypíše hodnotu, jinak vypíše "Věk není uveden".

```python
student = {
    "jméno": "Jan",
    "třída": "3.A"
}

# Doplň podmínku zde
```

```python
"vek" in student
```

```python
if "vek" in student:
    print(student["vek"])
else: 
    print("není uveden")
```

```python
# pridam vek do slovnika
student["vek"] = 25
```

```python
# ted uz vypise hodnotu
if "vek" in student:
    print(student["vek"])
else: 
    print("není uveden")
```

## Úloha 3: Filtrování seznamu

Máš seznam čísel. Vytvoř nový seznam, který obsahuje pouze čísla větší než 10.

```python
cisla = [5, 12, 3, 18, 7, 21, 9, 15]
# Vytvoř nový seznam zde
```

```python
# vytvorim si novy prazdny zoznam
# vo for prechadzam zadany 
# ak prvok je > 10, pridam ho do toho noveho zoznamu
```

```python
cisla = [5, 12, 3, 18, 7, 21, 9, 15]
nove_cisla = []

for c in cisla:
    if c > 10:
        nove_cisla.append(c)

nove_cisla
```

```python
# mozno krajsie cez comprehension
```

```python
nove_com = [c for c in cisla if c > 10]

nove_com
```
