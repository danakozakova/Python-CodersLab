## Základní klávesové zkratky

V Jupyter Lab existují dva režimy:
- **Edit mode** (zelený okraj buňky) - upravujete obsah buňky
- **Command mode** (modrý okraj buňky) - pracujete s buňkami

### Přepínání mezi režimy

- `Enter` - vstup do Edit mode
- `Esc` - vstup do Command mode

### Důležité zkratky v Command mode

- `A` - přidat buňku **nad** aktuální buňku (Above)
- `B` - přidat buňku **pod** aktuální buňku (Below)
- `M` - změnit buňku na Markdown
- `Y` - změnit buňku na Code
- `C` - kopírovat buňku (Copy)
- `V` - vložit buňku (Paste)
- `X` - vyjmout buňku (Cut)
- `DD` - smazat buňku (Delete - stiskněte D dvakrát)
- `Z` - vrátit zpět poslední příkaz (Undo)

### Spuštění buňky

- `Shift + Enter` - spustit buňku a přejít na další
- `Ctrl + Enter` - spustit buňku a zůstat v ní
