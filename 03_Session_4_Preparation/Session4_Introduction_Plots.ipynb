## 1. Grafické formy prezentace dat

V praxi se používá mnoho typů statistických grafů. Výběr konkrétního typu závisí na:
- povaze analyzovaného jevu
- typu dat
- charakteru zákonitostí, které má graf ukázat

### Nejčastěji používané typy grafů:

**Plošné grafy** (sloupcové, koláčové)
- Sloupcové a pruhové grafy - nejčastěji pro prezentaci struktury souboru v procentech
- Koláčové grafy - prezentace struktury souboru, začíná se na 12 hodinách, další výseče ve směru hodinových ručiček
- 3D grafy - prostorové varianty (kvádr místo obdélníku, válec místo kruhu)

**Čárové grafy**
- Linie vykreslená v souřadnicovém systému
- Nejčastěji pro prezentaci změn v čase

**Mapové grafy (kartogramy)**
- Ukazují variaci analyzovaných jevů pomocí barev
- Příklad: míra nezaměstnanosti, HDP na obyvatele

**Kombinované grafy**
- Kombinace různých typů (např. kartogram + sloupcový graf)
- Umožňují ukázat strukturu jevů v čase

**Radarové grafy**
- Vhodné pro porovnání několika stejně významných proměnných současně

## 3. Příklady kódu

### Nejjednodušší graf - pouze seznam čísel

````python
import matplotlib.pyplot as plt

plt.plot([0, 2, 4, 6])
plt.show()
````

### Neinteraktivní graf s popisky

````python
import matplotlib.pyplot as plt

y1 = [0, 1, 2, 3, 4, 5]
y2 = [0, 1, 4, 9, 16, 25]
````

````python
plt.plot(y1)
plt.plot(y2)
plt.show()
````

````python
plt.plot(y1, label= 'krivka1')
plt.plot(y2, label = 'krivka2')

plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")

plt.legend()
plt.show()
````

````python
plt.plot(y1, label='Plot 1')
plt.plot(y2, label='Plot 2')

plt.xlabel('x label')
plt.ylabel('y label')

plt.title("Simple Plot")
plt.legend()

plt.show()
````

### Interaktivní graf

Pro interaktivní grafy (zoom, posun, ukládání) použijte magic příkaz před kódem:

```python
%matplotlib ipympl
```

**Poznámka:** Pokud `ipympl` není nainstalován, nainstalujte ho přes Anaconda Navigator.

### Co znamená %matplotlib ipympl?
Příkaz `%matplotlib ipympl` je tzv. **"magic command"** (magický příkaz) a slouží na nastavení zpusobu, jakým se vykreslují grafy z knižnice **Matplotlib** vo vašem interaktivním prostředí.

* `%matplotlib`: Je to samotný magický příkaz, ktorý prepína backend (systém vykreslování) pro Matplotlib.
* `ipympl`: Je to specifický backend, ktorý umožňuje interaktivní vykreslování grafu.

**Interaktívne grafy**

Když použijete `plt.show()` bez tohto magic příkazu, dostanete statický obrázok grafu.

Když použijete `%matplotlib ipympl` (alebo starší `%matplotlib notebook`), Matplotlib použije interaktivní backend, ktorý vám přímo v okně notebooku umožní:
* Zoomovat (přibližovat) konkrétní časti grafu.
* Posouvat (panorámovat) graf do stran.
* Ukládat graf pomocí vstavaných nástroju přímo z jeho vokna.

## Zhrnutí

### Import knihovny
```python
import matplotlib.pyplot as plt
```

### Základní funkce
- `plt.plot(x, y)` - vykreslí čárový graf
- `plt.xlabel('text')` - popis osy x
- `plt.ylabel('text')` - popis osy y
- `plt.title('text')` - název grafu
- `plt.legend()` - zobrazí legendu
- `plt.show()` - zobrazí graf

### Backendy
Matplotlib používá různé backendy pro vykreslování grafů. Po zobrazení grafu obvykle nelze backend změnit. Pokud uvidíte hlášku o nemožnosti změny backendu, zavřete a znovu otevřete notebook.
