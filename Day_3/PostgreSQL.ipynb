```python
from psycopg2 import connect
```

```python
import getpass

password = getpass.getpass("Zadaj heslo: ")
print("Heslo bolo zadané.")
```

```python
cnx = connect(
    dbname="postgres",      # název vlastní DB
    user="postgres",
    password=password,
    host="localhost",       # nebo "127.0.0.1"
    port=5432
)
```

```python
cursor = cnx.cursor()
```

```python
cursor.execute('SELECT datname FROM pg_database;')
```

```python
import csv

with open('file.csv', mode='w') as csv_file:
    users = csv.writer(csv_file, delimiter=';', quotechar='"')
    for row in cursor:
        users.writerow(row)
```

```python
cnx.close()
```
