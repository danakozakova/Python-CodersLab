# ÄŒÃ¡st 3: BonusovÃ¡ ÄÃ¡st - Validace vstupÅ¯

V praxi je dÅ¯leÅ¾itÃ© kontrolovat, zda funkce dostÃ¡vÃ¡ sprÃ¡vnÃ½ typ vstupnÃ­ch dat. Tato sekce ukazuje, jak mÅ¯Å¾eme validovat vstupy a oÅ¡etÅ™it chybovÃ© stavy.

## PÅ™Ã­klad 1: NÃ¡sobenÃ­ s validacÃ­ vstupÅ¯

RozÅ¡Ã­Å™Ã­me funkci `nasobeni()` o kontrolu, Å¾e oba parametry jsou ÄÃ­sla.

```python
def nasobeni_s_validaci(a, b):
    """VrÃ¡tÃ­ souÄin dvou ÄÃ­sel s validacÃ­ vstupÅ¯.
    
    Args:
        a: PrvnÃ­ ÄÃ­slo (int nebo float)
        b: DruhÃ© ÄÃ­slo (int nebo float)
        
    Returns:
        SouÄin a * b
        
    Raises:
        TypeError: Pokud a nebo b nenÃ­ ÄÃ­slo
    """
    # Kontrola, zda jsou oba parametry ÄÃ­sla
    if not isinstance(a, (int, float)):
        raise TypeError(f"Parametr 'a' musÃ­ bÃ½t ÄÃ­slo, ale je typu {type(a).__name__}")
    
    if not isinstance(b, (int, float)):
        raise TypeError(f"Parametr 'b' musÃ­ bÃ½t ÄÃ­slo, ale je typu {type(b).__name__}")
    
    return a * b
```

```python
# Test s platnÃ½mi hodnotami
print('PlatnÃ© vstupy:')
print('2 * 3 =', nasobeni_s_validaci(2, 3))
print('2.5 * 4 =', nasobeni_s_validaci(2.5, 4))
```

```python
# Test s neplatnÃ½mi hodnotami
print('\nNeplatnÃ© vstupy:')
try:
    print(nasobeni_s_validaci("2", 3))  # String mÃ­sto ÄÃ­sla
except TypeError as e:
    print(f"Chyba: {e}")

try:
    print(nasobeni_s_validaci(2, [3]))  # Seznam mÃ­sto ÄÃ­sla
except TypeError as e:
    print(f"Chyba: {e}")
```

### ğŸ” VysvÄ›tlenÃ­ kÃ³du:

- `isinstance(a, (int, float))` kontroluje, zda je `a` typu int nebo float
- `raise TypeError(...)` vyvolÃ¡ vÃ½jimku (chybu) s popisnou zprÃ¡vou
- `try-except` blok zachytÃ­ vÃ½jimku a umoÅ¾nÃ­ nÃ¡m ji zpracovat

## PÅ™Ã­klad 2: PÅ™evod mÄ›n s validacÃ­

RozÅ¡Ã­Å™Ã­me funkci `usd_na_czk()` o kontrolu, Å¾e ÄÃ¡stka je ÄÃ­slo a je kladnÃ¡.

```python
def usd_na_czk_s_validaci(usd):
    """PÅ™evede USD na CZK s validacÃ­ vstupÅ¯.
    
    Args:
        usd: ÄŒÃ¡stka v USD (musÃ­ bÃ½t kladnÃ© ÄÃ­slo)
        
    Returns:
        ÄŒÃ¡stka v CZK
        
    Raises:
        TypeError: Pokud usd nenÃ­ ÄÃ­slo
        ValueError: Pokud usd je zÃ¡pornÃ©
    """
    # Kontrola typu
    if not isinstance(usd, (int, float)):
        raise TypeError(f"ÄŒÃ¡stka musÃ­ bÃ½t ÄÃ­slo, ale je typu {type(usd).__name__}")
    
    # Kontrola hodnoty
    if usd < 0:
        raise ValueError(f"ÄŒÃ¡stka musÃ­ bÃ½t kladnÃ¡, ale je {usd}")
    
    kurz = 23
    return usd * kurz
```

```python
# Test s platnÃ½mi hodnotami
print('PlatnÃ© vstupy:')
print('100 USD =', usd_na_czk_s_validaci(100), 'CZK')
print('0 USD =', usd_na_czk_s_validaci(0), 'CZK')  # Nula je v poÅ™Ã¡dku
```

```python
# Test s neplatnÃ½mi hodnotami
print('\nNeplatnÃ© vstupy:')
try:
    print(usd_na_czk_s_validaci("100"))  # String mÃ­sto ÄÃ­sla
except TypeError as e:
    print(f"Chyba typu: {e}")

try:
    print(usd_na_czk_s_validaci(-50))  # ZÃ¡pornÃ¡ hodnota
except ValueError as e:
    print(f"Chyba hodnoty: {e}")
```

### ğŸ” VysvÄ›tlenÃ­ rozdÃ­lu TypeError vs ValueError:

- **TypeError**: Å patnÃ½ typ dat (napÅ™. string mÃ­sto ÄÃ­sla)
- **ValueError**: SprÃ¡vnÃ½ typ, ale neplatnÃ¡ hodnota (napÅ™. zÃ¡pornÃ© ÄÃ­slo)

## PÅ™Ã­klad 3: Odmocnina s validacÃ­

RozÅ¡Ã­Å™Ã­me funkci `odmocnina()` o kontrolu, Å¾e ÄÃ­slo nenÃ­ zÃ¡pornÃ© (reÃ¡lnÃ¡ odmocnina ze zÃ¡pornÃ©ho ÄÃ­sla neexistuje).

```python
def odmocnina_s_validaci(cislo, stupen=2):
    """VypoÄÃ­tÃ¡ stupen-tou odmocninu z ÄÃ­sla s validacÃ­.
    
    Args:
        cislo: ÄŒÃ­slo, ze kterÃ©ho poÄÃ­tÃ¡me odmocninu (musÃ­ bÃ½t nezÃ¡pornÃ©)
        stupen: StupeÅˆ odmocniny (vÃ½chozÃ­ 2)
        
    Returns:
        Odmocninu
        
    Raises:
        TypeError: Pokud cislo nebo stupen nenÃ­ ÄÃ­slo
        ValueError: Pokud cislo je zÃ¡pornÃ© nebo stupen je 0
    """
    # Kontrola typÅ¯
    if not isinstance(cislo, (int, float)):
        raise TypeError(f"Parametr 'cislo' musÃ­ bÃ½t ÄÃ­slo, ale je typu {type(cislo).__name__}")
    
    if not isinstance(stupen, (int, float)):
        raise TypeError(f"Parametr 'stupen' musÃ­ bÃ½t ÄÃ­slo, ale je typu {type(stupen).__name__}")
    
    # Kontrola hodnot
    if cislo < 0:
        raise ValueError(f"Nelze poÄÃ­tat reÃ¡lnou odmocninu ze zÃ¡pornÃ©ho ÄÃ­sla ({cislo})")
    
    if stupen == 0:
        raise ValueError("StupeÅˆ odmocniny nemÅ¯Å¾e bÃ½t 0")
    
    return cislo ** (1 / stupen)
```

```python
# Test s platnÃ½mi hodnotami
print('PlatnÃ© vstupy:')
print('2. odmocnina z 64 je', odmocnina_s_validaci(64, 2))
print('2. odmocnina z 64 je', odmocnina_s_validaci(64))  # vÃ½chozÃ­ hodnota
print('4. odmocnina z 2401 je', odmocnina_s_validaci(2401, 4))
```

```python
# Test s neplatnÃ½mi hodnotami
print('\nNeplatnÃ© vstupy:')
try:
    print(odmocnina_s_validaci(-64, 2))  # ZÃ¡pornÃ© ÄÃ­slo
except ValueError as e:
    print(f"Chyba hodnoty: {e}")

try:
    print(odmocnina_s_validaci(64, 0))  # NultÃ½ stupeÅˆ
except ValueError as e:
    print(f"Chyba hodnoty: {e}")

try:
    print(odmocnina_s_validaci("64", 2))  # String mÃ­sto ÄÃ­sla
except TypeError as e:
    print(f"Chyba typu: {e}")
```

### Å˜eÅ¡enÃ­ cviÄenÃ­ s validacÃ­

```python
def deleni_s_validaci(delenec, delitel):
    """VydÄ›lÃ­ delenec dÄ›litelem s validacÃ­ vstupÅ¯.
    
    Args:
        delenec: ÄŒÃ­slo, kterÃ© dÄ›lÃ­me
        delitel: ÄŒÃ­slo, kterÃ½m dÄ›lÃ­me (nesmÃ­ bÃ½t 0)
        
    Returns:
        VÃ½sledek dÄ›lenÃ­
        
    Raises:
        TypeError: Pokud parametry nejsou ÄÃ­sla
        ValueError: Pokud je dÄ›litel 0
    """
    # Kontrola typÅ¯
    if not isinstance(delenec, (int, float)):
        raise TypeError(f"DÄ›lenec musÃ­ bÃ½t ÄÃ­slo, ale je typu {type(delenec).__name__}")
    
    if not isinstance(delitel, (int, float)):
        raise TypeError(f"DÄ›litel musÃ­ bÃ½t ÄÃ­slo, ale je typu {type(delitel).__name__}")
    
    # Kontrola hodnoty
    if delitel == 0:
        raise ValueError("Nelze dÄ›lit nulou!")
    
    return delenec / delitel
```
