## 3. Množiny (Sets)

#### Set (Množina) vs Dictionary (Slovník)

##### Set (množina) - pouze hodnoty
* Obsahuje jen hodnoty (bez klíčů)
* Nemá duplicity
* Nemá pořadí (neseřazené)

```python
pythonbarvy = {'červená', 'zelená', 'modrá'}
cisla = {1, 2, 3, 4, 5}

print(type(pythonbarvy))
print(type(cisla))
```

##### Dictionary (slovník) - páry klíč: hodnota

* Obsahuje páry klíč: hodnota (s dvojtečkou!)
* Klíče jsou unikátní
* Od Pythonu 3.7+ má pořadí

```python
osoba = {'jmeno': 'Anna', 'vek': 25, 'mesto': 'Praha'}
print(type(osoba))
```

#### POZOR na prázdné závorky!
Prázdné `{}` Python vždycky považuje za **slovník**, ne množinu!

```python
prazdny = {}
print(type(prazdny))
```

```python
prazdna = set() 
print(type(prazdna))
```

### Průnik (intersection) - společné prvky ∩

* Metoda `intersection()` nebo operátor `&`

```python
mnozina1 = {1, 2, 3, 4, 5}
mnozina2 = {4, 5, 6, 7, 8}

# Způsob 1: Metoda
prunik = mnozina1.intersection(mnozina2)
print(prunik) 

# Způsob 2: Operátor &
prunik = mnozina1 & mnozina2
print(prunik) 
```

### Sjednocení (union) - všechny prvky ∪

* Metoda `union()` nebo operátor `|`

```python
mnozina1 = {1, 2, 3, 4, 5}
mnozina2 = {4, 5, 6, 7, 8}

# Způsob 1: Metoda
sjednoceni = mnozina1.union(mnozina2)
print(sjednoceni)  # {1, 2, 3, 4, 5, 6, 7, 8}

# Způsob 2: Operátor |
sjednoceni = mnozina1 | mnozina2
print(sjednoceni)  # {1, 2, 3, 4, 5, 6, 7, 8}
```
