# Cvičení 1: Práce s knihovnami

Máme číslo `n = 5`. Vypočítej faktoriál tohoto čísla pomocí funkce `factorial` z knihovny `math`.

**Připomenutí:** Faktoriál 5! = 5 × 4 × 3 × 2 × 1 = 120

Doplň chybějící import a správný způsob volání funkce:

```python
import math
```

```python
# Doplň správný import knihovny math

n = 5
# Doplň správné volání funkce factorial
faktorial = math.factorial(n)

print(f"Faktoriál čísla {n} je {faktorial}")
```

Existují dva způsoby, jak importovat funkci `factorial`:

1. `import math` → všechny funkce jsou dostupný cez název modulu, pak používáme `math.factorial()`
2. `from math import factorial` → importujeme jenom funkci factorial, pak používáme přímo `factorial()`

Vytvoř program, který:
1. Použije knihovnu `random` (nemusí se instalovat, je součástí Pythonu)
2. Vygeneruje 5 náhodných čísel od 1 do 10
3. Pro každé číslo vypočítá a vypíše faktoriál

**Nápověda:** Použij `random.randint(1, 10)` pro generování náhodných čísel

```python
# Napiš zde celý kód
import random
```

```python
random.randint(1,10)
```

```python
# Vygeneruje 5 náhodných čísel od 1 do 10
```

```python
random.randint(1,10) * 5
```

```python
for i in range(5):
    cislo = random.randint(1,10)
    print(cislo)
```

```python
# Pro každé číslo vypočítá a vypíše faktoriál
for i in range(5):
    cislo = random.randint(1,10)
    print(math.factorial(cislo))
```

```python
for i in range(5):
    cislo = random.randint(1,10)
    print(f"{cislo}! = {math.factorial(cislo)}")
```
