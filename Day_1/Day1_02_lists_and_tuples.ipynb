## 2. Pokroƒçil√© indexov√°n√≠ - slicing (≈ôezy)

### Teorie
- `seznam[start:end]` - od start do end (bez end!)
- `seznam[:cislo]` - od zaƒç√°tku do ƒç√≠slo
- `seznam[cislo:]` - od ƒç√≠slo do konce
- `seznam[start:end:krok]` - s krokem

**Pozor:** Rozsah je pravostrannƒõ otev≈ôen√Ω!

```python
# P≈ô√≠klad
bbt = ["01Leonard", "02Sheldon", "03Penny", "04Howard", "05Raj", "06Bernardette", "07Amy"]
```

```python
print(bbt[0:1])    # ['Leonard']
```

```python
print(bbt[2:5])    # ['Penny', 'Howard', 'Raj']
```

```python
print(bbt[4:])     # ['Raj', 'Bernardette', 'Amy']
```

```python
print(bbt[:2])     # ['Leonard', 'Sheldon']
```

```python
print(bbt[:-4])    # ['Leonard', 'Sheldon', 'Penny']
```

```python
print(bbt[-3:])    # ['Raj', 'Bernardette', 'Amy']
```

## 6. List Comprehension

### Teorie
List comprehension je elegantn√≠ zp≈Øsob vytvo≈ôen√≠ seznamu.

**Syntaxe:**
- Z√°kladn√≠: `[vyraz for prvek in kolekce]`
- S podm√≠nkou: `[vyraz for prvek in kolekce if podminka]`

```python
# P≈ô√≠klad: Sud√° ƒç√≠sla od 0 do 20
# Klasick√Ω zp≈Øsob:
my_list = []
for i in range(0, 11):
    my_list.append(i * 2)
print(my_list)
```

```python
# S list comprehension:
my_list = [2 * i for i in range(0, 11)]
print(my_list)
```

```python
# P≈ô√≠klad s podm√≠nkou: ƒç√≠sla dƒõliteln√° t≈ôemi
my_list = [i for i in range(0, 18) if i % 3 == 0]
print(my_list)  # [0, 3, 6, 9, 12, 15]
```
